# Statistical Summary

```python
import pandas as pd
import numpy as np

# Example dataset
data = pd.read_csv(r"C:\Users\Lenovo\Desktop\Cardiotocographic.csv")
df = pd.DataFrame(data)
df
```

```python
# Summary statistics
summary = df.describe()
summary
```

# Data Visualization:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
# Create histograms for numerical variables
data.hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Variables')
plt.show()
```

```python
# Create boxplots for numerical variables
plt.figure(figsize=(8,4))
sns.boxplot(data=data)
plt.title('Boxplots of Numerical Variables')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar charts for categorical variables
categorical_columns = ['NSP', 'Tendency'] 
for col in categorical_columns:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, data=data)
    plt.title(f'Bar Chart of {col}')
    plt.show()
```

```python
# Pie charts for categorical variables
for col in categorical_columns:
    plt.figure(figsize=(4,4))
    data[col].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')
    plt.show()
```

```python
# Scatter plots for exploring relationships
plt.figure(figsize=(4,4))
sns.scatterplot(x='LB', y='AC', data=data)  # Replace with actual numerical variable names
plt.title('Scatter Plot between Numerical LB and AC')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,4))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pair plots for deeper insights
sns.pairplot(data)
plt.title('Pair Plot of Variables')
plt.show()
```

```python
 # Violin plots for deeper insights
plt.figure(figsize=(10, 6))
sns.violinplot(x='Tendency', y='AC', data=data)  # Replace with actual variable names
plt.title('Violin Plot of Numerical Var by Categorical Var')
plt.show()
```
